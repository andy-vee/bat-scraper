```python
# Separate Status from listing title
bat1['Listing Status'] = [bat1['Listing Title'][i].split(' - ', maxsplit=1)[1] for i in range(len(bat1))]
bat1['Listing Status'] = [bat1['Listing Status'][i].split(' (Lot',maxsplit=1)[0] for i in range(len(bat1))]
```

```python
#Generate Clean listing title
bat1['Clean Title'] = [bat1['Listing Title'][i].split(' - ',maxsplit=1)[0] for i in range(len(bat1))]
bat1['Clean Title'] = [bat1['Listing Title'][i].split(' for sale',maxsplit=1)[0] for i in range(len(bat1))]
```

```python
# Generate factor-based listing status variable (open, closed)

for row in range(len(bat1)):
    bat1['Open'][row] = False
    if bat1['Listing Status'][row].split(' ')[0] == "ending":
        bat1['Open'][row] = True
    elif bat1['Listing Status'][row].split(' ')[0] == "closed":
        bat1['Open'][row] = False
```

```python
#Generate integer variable for # of viewers, watchers, comments
bat1['Num_Viewers']=[str(bat1['View Stats'][i]).split(' views')[0] for i in range(len(bat1))]
for i in range(len(bat1)):
    if (type(bat1['Num_Viewers'][i]) == str) and (bat1['Num_Viewers'][i] != 'nan'):
        bat1['Num_Viewers'][i] = int(bat1['Num_Viewers'][i].replace(',',''))
    elif (bat1['Num_Viewers'][i] == 'nan'):
        bat1['Num_Viewers'][i] = 0
```

```python
#Generate integer variable for # of watchers
# bat1['Num_Watchers']=[str(bat1['View Stats'][i]).split(' views,')[1] for i in range(len(bat1))]
# bat1['Num_Watchers']=[bat['Num_Watchers'][i].replace(' watchers','') for i in range(len(bat1))]
bat1['Num_Watchers'] = ['blank' for i in range(len(bat1))]
for i in range(len(bat1)):
    if (type(bat1['View Stats'][i]) == str) and (bat1['View Stats'][i] != 'nan'):
        bat1['Num_Watchers'][i] = bat1['View Stats'][i].split(' views,')[1]
        bat1['Num_Watchers'][i] = bat1['Num_Watchers'][i].replace(' watchers','')
        bat1['Num_Watchers'][i] = int(bat1['Num_Watchers'][i].replace(',',''))
    elif (bat1['Num_Watchers'][i] == 'nan') or (bat1['Num_Watchers'][i] == 'NaN') or (bat1['Num_Watchers'][i]=="blank"):
        bat1['Num_Watchers'][i] = 0
```
